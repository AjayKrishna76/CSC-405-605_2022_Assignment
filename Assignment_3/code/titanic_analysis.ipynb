*** 3. Create age categories: children (under 14 years), adolescents (14-20), adult (21-64), and senior(65+), and calculate survival proportions by age category, class and sex. (40 pts)***

```python
# Creating age categories (seperating the given dataset according to the different age groups)
children = t_df.loc[t_df['age'] < 14]
adolescents = t_df.loc[(t_df['age'] >= 14) & (t_df['age'] < 21)]
adult = t_df.loc[(t_df['age'] >= 21) & (t_df['age'] < 65)]
senior = t_df.loc[t_df['age'] >= 65]
```

```python
children
```

creating a column age category which has the catogery of the pasengor with respect to age  

```python
t_df
```

```python
t_df["age_category"] = ""
t_df
```

```python
print(len(t_df))
```

```python
#with the help of while loop reading the data and assigning the category for all the rows
i = 1
while (i<1309):
    if (t_df.age[i] < 14.0):
        t_df.age_category[i] = "child"
    elif (t_df.age[i] >= 14.0) and (t_df.age[i] < 21.0):
        t_df.age_category[i] = "adolescent"
    elif (t_df.age[i] >= 21.0) and (t_df.age[i] < 65.0):
        t_df.age_category[i] = "adult"
    elif (t_df.age[i] >= 65.0):
        t_df.age_category[i] = "senior"
    else:
        t_df.age_category[i] = "NaN"
    i+=1
```
